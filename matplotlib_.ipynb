```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# ------------Figure and Subplot-----------------
```

```python

# 图像位于Figure对象中
fig = plt.figure(figsize=(10,8))
# 在Figure对象添加subplot(2行2列)
ax1 = fig.add_subplot(2,2,1)  # 第一个
ax1.plot(np.arange(15).cumsum(),"k--")
ax2 = fig.add_subplot(2,2,2)  # 第二个 
ax2.hist(np.random.rand(100),bins=20,color='k',alpha = 0.3)
ax3 = fig.add_subplot(2,2,3)  # 第三个
ax3.scatter(np.arange(30),np.random.rand(30))
plt.show()
```

```python
# 直接返回Figure对象和subplot的数组（axes）
fig,axes = plt.subplots(2,2,figsize=(10,8)) 
# 直接通过下标索引
axes[0,0].plot(np.arange(15).cumsum(),"k--")                      # 第一个，线性图
axes[0,1].hist(np.random.rand(100),bins=20,color='k',alpha = 0.3) # 第二个，直方图
axes[1,1].scatter(np.arange(30),np.random.rand(30))               # 第四个，散点图
plt.subplots_adjust(wspace=0,hspace=0.2)                          # 调整间距
plt.show()
```

```python
# ----------颜色、标记、线型--------------------
```

```python
x = np.random.randn(30).cumsum()
plt.plot(x,'ko--') #颜色必须在前面
plt.show()
```

```python
# 标记marker:强调实际数据点
# linestyle:线型
# drawstyle:插值的方式---默认的是连续的线性方式
plt.plot(x,color="k",linestyle="--",drawstyle="steps-post",marker="o")
plt.show()
```

```python
# ----------------刻度，标签，图例----------------
```

```python
fig = plt.figure(figsize=(20,8))
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum(),'k*',label="one")
ax.plot(np.random.randn(1000).cumsum(),'g-',label="two")
ax.plot(np.random.randn(1000).cumsum(),'b.',label="three")
ax.set_title("A example of plot")         # 标题 
ax.set_xlabel("X")                        # x轴标签
ax.set_xticks([0,250,500,750,1000])       # x轴刻度
ax.set_xticklabels(["a",'b','c','d','e'],rotation = 30,fontsize = "small") # 修改x轴刻度的标签，rotation旋转，fontsize字体大小
ax.set_ylabel("Y")                        # y轴标签
ax.set_ylim([-60,60])                     # y轴范围
ax.legend(loc='best')                     # 显示legend图例，位置best=最适合的位置
plt.show()
```

```python
# 对于plt,是x()方法。对应于subplot对象的set_x()方法
fig1 = plt.figure(figsize=(20,8))
plt.plot(np.random.randn(1000).cumsum(),'k*',label="one")
plt.plot(np.random.randn(1000).cumsum(),'g-',label="two")
plt.plot(np.random.randn(1000).cumsum(),'b.',label="three")
plt.title("A example of plot")         # 标题 
plt.xlabel("X")                        # x轴标签
plt.xticks([0,250,500,750,1000],rotation = 30,fontsize = "small")       # x轴刻度,rotation旋转，fontsize字体大小
plt.ylabel("Y")                        # y轴标签
plt.ylim([-60,60])                     # y轴范围
plt.legend(loc='best')                 # 显示legend图例，位置best=最适合的位置
plt.show()
```

```python
# -----------------pandas中的绘图函数--------------------
```

```python
# 1.线型图
from pandas import Series
# Series
s = Series(np.random.randn(10),index=np.arange(0,100,10))
s.plot(kind='line',color = 'b',style='-.',marker='o',alpha=0.5,label="Serise",
      use_index=True,rot=30,ylim=[-2,2],yticks=[-2,-1,0,1,2],grid = True)
# kind 图的类型，line(线型)，bar(柱状图)，barh(水平柱状图)，kde(密度图)---------默认为线型
# alpha不透明度0-1
# use_index 将对象的索引作为刻度标签
# rot旋转刻度标签0-360
# xticks ,yticks刻度值[1,3,5,7...]
# xlim,ylim范围[0,10]
# grid 显示网格
plt.legend(loc="best")
plt.show()
```

```python
from pandas import DataFrame
df = DataFrame(np.random.randn(10,4),columns=['a','b','c','d'],index=np.arange(1,11,1))
df.plot() 
plt.show()
```

```python
# DataFrame
from pandas import DataFrame
df = DataFrame(np.random.randn(10,4),columns=['a','b','c','d'],index=np.arange(1,11,1))
df.plot(subplots=True) #subplots将数据绘制到单独的subplot
plt.show()
```

```python
# 2.柱状图，kind=bar、barh
# Series
fig,axes = plt.subplots(2,1)
data = Series(np.random.rand(10),index = list('abcdefghij'))
# 通过ax指定subplot对象
data.plot(kind='bar',ax=axes[0],color='k')
data.plot(kind='barh',ax=axes[1],color='b')
plt.show()
```
