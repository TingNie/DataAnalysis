```python
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# 返回目录中的文件名列表
import os
def get_imlist(path):
    path_imlist = [os.path.join(path,f) for f in os.listdir(path) if f.endswith('.jpg')]
    imlist = [f for f in os.listdir(path) if f.endswith('.jpg')]
    return path_imlist,imlist
path_imlist,imlist = get_imlist('image')
print(path_imlist)
print(imlist)
```

```python
pil_im = Image.open('image/my.jpg').convert('L') # 颜色转化，灰度化
plt.imshow(pil_im)
```

```python
pil_im.thumbnail((128,128)) # 缩略图
plt.imshow(pil_im)
```

```python
#拷贝粘贴区域
pil_im = Image.open('image/my.jpg')
box = (600,400,900,600) # 左上角（0,0），左，上，右，下
region = pil_im.crop(box)
region = region.transpose(Image.ROTATE_180)
pil_im.paste(region,box)
plt.imshow(pil_im)
```

```python
# 调整图像尺寸
pil_im = Image.open('image/my.jpg')
size = (128,128)
pil_im = pil_im.resize(size)
print pil_im.size

# 旋转图像45°
pil_im = pil_im.rotate(45)
plt.imshow(pil_im)
```

```python
im = np.array(Image.open('image/my.jpg'))# 读取图像到数组
x = [100, 100, 400, 400]
y = [200, 500, 200, 500]
plt.plot(x, y, 'r*')
plt.plot(x[:2], y[:2])
plt.imshow(im)
```

```python
#图像的轮廓信息
im = np.array(Image.open('image/my.jpg').convert("L"))# 读取图像到数组
plt.figure()
plt.gray()#不使用颜色信息
plt.contour(im,origin='image') # 再原点的左上角显示轮廓图像
plt.imshow(im)
```

```python
#图像的直方图
im = np.array(Image.open('image/my.jpg').convert("L"))# 读取图像到数组
plt.figure()
plt.hist(im.flatten(),128)
plt.show()
```

```python
# 灰度转化
im = np.array(Image.open('image/my.jpg').convert("L"))# 读取图像到数组
im2 = 255 - im
plt.subplot(131)
plt.imshow(im2)
im3 = (100.0/255)*im + 100
plt.subplot(132)
plt.imshow(im3)
im4 = 255.0 * (im/255.0)**2
plt.subplot(133)
plt.imshow(im4)
```
